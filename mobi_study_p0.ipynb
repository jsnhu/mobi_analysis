```python
import numpy as np
import sqlite3
from bs4 import BeautifulSoup 
import matplotlib.pyplot as plt
import requests
import regex as re
import os
```

```python
mobi_data_url = "https://www.mobibikes.ca/en/system-data"
mobi_page = requests.get(mobi_data_url)
mobi_page_soup = BeautifulSoup(mobi_page.content, 'html.parser')
```

```python
links = mobi_page_soup.find_all('a', href=True)

id_dict = {}
regex = r'''d/(.*?)/view'''


for link in links:
    href = link['href']
    if "drive.google" in href:
        date = link.get_text()
        date = date.replace(" ", "_")
        match = re.search(regex, href)
        if match:
            file_id = match.group(1)
            id_dict[date] = file_id
            print(f"Adding Date: {date}, File ID: {file_id}")
        else:
            print("******************************No match found.")
            
```

```python
id_dict.pop("ALL_of_2017")
id_dict['ALL_of_2017'] = id_dict.pop("\xa0")
```

```python
def construct_download_url(file_id):
    base_front_url = "https://drive.usercontent.google.com/u/0/uc?id="
    base_back_url = "&export=download"
    return base_front_url + file_id + base_back_url
```

```python
test_filename = list(id_dict.keys())[1]
file_path = os.path.join("mobi_csv", f"{test_filename}.csv")
with open(file_path, 'wb') as file:
    download_url = construct_download_url(id_dict[test_filename])
    response = requests.get(download_url)
    file.write(response.content)
```
